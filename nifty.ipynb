```python
sector_data = pd.read_csv('sector_analysis.csv')
```

```python
# Step 1 & 2: Calculate the combined financial sector contributions
financial_sector_rows = sector_data[sector_data['Sector'].isin(['BanksPV', 'BanksPS', 'NBFC'])]
financial_sector_sum = financial_sector_rows.drop('Sector', axis=1).sum()

# Step 3: Create a new dataframe row for the combined financial sector
financial_sector_combined = pd.DataFrame([financial_sector_sum], columns=financial_sector_rows.columns.drop('Sector'))
financial_sector_combined['Sector'] = 'Financial Sector'

# Step 4: Remove the original financial sector rows
sector_data_excluded = sector_data[~sector_data['Sector'].isin(['BanksPV', 'BanksPS', 'NBFC'])]

# Step 5: Concatenate the modified original dataframe with the new financial sector row
# Ensuring 'Sector' is the first column
cols = ['Sector'] + [col for col in financial_sector_combined.columns if col != 'Sector']
financial_sector_combined = financial_sector_combined[cols]

sector_data_combined = pd.concat([sector_data_excluded, financial_sector_combined], ignore_index=True)

# Display the tail of the updated dataframe to verify the changes
sector_data_combined.tail()
```

```python
def plot_sector_distribution(df, year):
    """
    Plots a pie chart for the distribution among the sectors for a specific year.

    Parameters:
    - df: DataFrame containing sector-wise contributions.
    - year: Specific year for which the distribution is to be plotted.

    The function assumes that the 'Sector' column contains sector names and that
    each other column represents a specific year with numerical values indicating
    the contribution percentages.
    """
    # Check if the year is a valid column in the dataframe
    if str(year) not in df.columns:
        print(f"Year {year} is not available in the dataset.")
        return
    
    # Prepare data for plotting
    sector_names = df['Sector']
    contributions = df[str(year)]

    # Plotting
    plt.figure(figsize=(10, 7))
    plt.pie(contributions, labels=sector_names, autopct='%1.1f%%', startangle=140)
    plt.title(f'Sector Distribution in {year}')
    plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
    plt.show()
```

```python
def plot_sector_trend_over_years(df, sector, years):
    """
    Plots a line graph showing the trend of a sector's contribution over selected years.

    Parameters:
    - df: DataFrame containing sector-wise contributions.
    - sector: The sector for which the trend is to be plotted.
    - years: List of years for which the trend is to be plotted.
    
    The function assumes that the 'Sector' column contains sector names and that
    each other column represents a specific year with numerical values indicating
    the contribution percentages.
    """
    # Filter the dataframe for the specified sector
    sector_data = df[df['Sector'] == sector]
    
    # Check if the sector exists in the dataframe
    if sector_data.empty:
        print(f"Sector '{sector}' not found in the dataset.")
        return
    
    # Extract the contributions for the specified years
    contributions = sector_data[years].values.flatten()
    
    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(years, contributions, marker='o', linestyle='-', color='b')
    plt.title(f'Trend of {sector} Sector Contribution (2002-2024)')
    plt.xlabel('Year')
    plt.ylabel('Contribution Percentage')
    plt.grid(True)
    plt.xticks(years, years)  # Ensure all specified years are shown as ticks
    plt.show()


```
